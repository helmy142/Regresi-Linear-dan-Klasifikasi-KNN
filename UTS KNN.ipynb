```python
data=data.drop('time_left',axis=1) # untuk menghapus atribut yang tidak memengaruhi data class
```

## MENGUBAH OBJECT MENJADI NUMERIK

```python
from sklearn.preprocessing import LabelEncoder 
for col in ['map','bomb_planted','round_winner']:#karena data pada atribut map, bomb_planted dan round winner bukan numerik
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    data
    print (data['map'],data['bomb_planted'],data['round_winner'])
```

## DATA X DAN Y

```python
x = data.drop('round_winner', axis = 1)#Pengahpusan atribut ini dikarenakan atribut tersebut bukan merupakan variable yang memengaruhi data class
y = data['round_winner']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state =0) 
```
